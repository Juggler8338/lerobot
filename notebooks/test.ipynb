```python
import gymnasium as gym
import lerobot
import numpy as np
import os

# 强制设置环境变量 (为了保险)
os.environ["MUJOCO_GL"] = "egl"

print("正在初始化环境...")
try:
    # 使用 LeRobot 里的一个简单环境，或者原生 MuJoCo 环境
    # 这里以 Aloha 环境为例 (如果已安装) 或者通用的 gym 环境
    env = gym.make("Ant-v4", render_mode="rgb_array") 
    
    obs, _ = env.reset()
    print("环境重置成功。")
    
    # 测试渲染
    img = env.render()
    print(f"渲染成功，图像形状: {img.shape}") 
    
    # 如果能走到这一步，说明 Headless 渲染没问题
    print("Headless 环境验证通过！")
    
except Exception as e:
    print(f"报错了: {e}")
    print("提示：检查是否安装了 libglew 或类似依赖，或者显卡驱动是否正常。")
```
